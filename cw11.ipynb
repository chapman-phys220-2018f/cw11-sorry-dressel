```python
import ode
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(setVals, N, title1):
    plt.figure(figsize=(8,6))
    a = plt.axes()
    set1 = [x[0] for x in setVals]
    set2 = [x[1] for x in setVals]
    a.plot(np.arange(0, 10*np.pi, 2*np.pi/N), set1[0:-1], color= 'orange', label="cos(t)")
    a.plot(np.arange(0, 10*np.pi, 2*np.pi/N), set2[0:-1], color= 'yellow', label="-sin(t)")
    a.plot(N,set1[0:-1],color="k",label='e vs x')
    a.plot(N,set2[1,:-1],color="green",label='e vs y')
    a.set(xlabel = "t", ylabel="f(t)", title=title1)
    a.legend()
    a.grid
    plt.show()
```

```python
plot((ode.euler_2(700, np.array([1, 0]))), 700, "Euler Plot 1")
```

```python
plot((ode.euler_2(100, np.array([1, 0]))), 100, "Euler Plot 1")
```

```python
plot((ode.Heuns_2(1000, np.array([1, 0]))), 1000, "Euler Plot 1")
```
