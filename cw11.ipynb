```python
import ode
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(setVals, N, title1):
    plt.figure(figsize=(8,6))
    a = plt.axes()
    set1 = [x[0] for x in setVals]
    set2 = [x[1] for x in setVals]
    a.plot(np.arange(0, 10*np.pi, 2*np.pi/N), set1[0:-1], color= 'orange', label="cos(t)")
    a.plot(np.arange(0, 10*np.pi, 2*np.pi/N), set2[0:-1], color= 'blue', label="-sin(t)")
    a.set(xlabel = "t", ylabel="f(t)", title=title1)
    a.legend()
    a.grid
    plt.show()
```

```python
plot((ode.euler_2(700, np.array([1, 0]))), 700, "Euler Plot ")
```

In this one we see that the approximation is not super accurate, it has the shape of a reverse damped oscillation.

```python
plot((ode.Heuns_2(50, np.array([1, 0]))), 50, "Heun's Plot")
```

This one oscillates alot better than the euler method. For the most part the oscillations are near to perfect.

```python
plot((ode.rungeKuttaSecond_2(30, np.array([1, 0]))), 30, "Runge Kutta Second")
```

```python
plot((ode.rungeKuttaFourth_2(9, np.array([1, 0]))), 9, "Runge Kutta Fourth")
```

As you go through the methods, we see that the smallest nth value gets smaller and smaller. Showing that the methods get more accurate one by one.
