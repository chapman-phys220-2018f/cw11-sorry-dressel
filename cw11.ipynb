```python
import ode
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(setVals, N, title1):
    plt.figure(figsize=(8,6))
    a = plt.axes()
    set1 = [x[0] for x in setVals]
    set2 = [x[1] for x in setVals]
    a.plot(np.arange(0, 10*np.pi, 2*np.pi/N), set1[0:-1], color= 'orange', label="cos(t)")
    a.plot(np.arange(0, 10*np.pi, 2*np.pi/N), set2[0:-1], color= 'blue', label="-sin(t)")
    a.set(xlabel = "t", ylabel="f(t)", title=title1)
    a.legend()
    a.grid
    plt.show()
```

```python
plot((ode.euler_2(100, np.array([1, 0]))), 100, "Euler Plot ")
```

```python
plot((ode.euler_2(700, np.array([1, 0]))), 700, "Euler Plot ")
```

```python
plot((ode.euler_2(1000, np.array([1, 0]))), 1000, "Euler Plot ")
```

In this one we see that the approximation is not super accurate, it has the shape of a reverse damped oscillation, as we raise the n value it gets more accurate and seems to be very accurate around a value of 1000

```python
plot((ode.Heuns_2(10, np.array([1, 0]))), 10, "Heun's Plot")
```

```python
plot((ode.Heuns_2(50, np.array([1, 0]))), 50, "Heun's Plot")
```

```python
plot((ode.Heuns_2(100, np.array([1, 0]))), 100, "Heun's Plot")
```

This one oscillates alot better than the euler method. For the most part the oscillations are near to perfect and only needs an n value of 100 to get this perfect oscillation, thats 1/10 of the euler method

```python
plot((ode.rungeKuttaSecond_2(10, np.array([1, 0]))), 10, "Runge Kutta Second")
```

```python
plot((ode.rungeKuttaSecond_2(30, np.array([1, 0]))), 30, "Runge Kutta Second")
```

```python
plot((ode.rungeKuttaSecond_2(100, np.array([1, 0]))), 100, "Runge Kutta Second")
```

This one got alot better than the heuns method and so much better than the euler|, it works with a value of 30 and has a very good oscillation

```python
plot((ode.rungeKuttaFourth_2(10, np.array([1, 0]))), 10, "Runge Kutta Fourth")
```

```python
plot((ode.rungeKuttaFourth_2(20, np.array([1, 0]))), 20, "Runge Kutta Fourth")
```

```python
plot((ode.rungeKuttaFourth_2(30, np.array([1, 0]))), 30, "Runge Kutta Fourth")
```

This one is super accurate representation of the oscillation we are looking for. It takes a lower value of n to give us an accurate value, but not too much better than the R2 method did

As you go through the methods, we see that the smallest nth value gets smaller and smaller. Showing that the methods get more accurate in the sequence shown. They also take less runs to be accurate. The euler method took about 1000 to give an accurate oscillation, where as the RK4 method only took between 20-30 runs
